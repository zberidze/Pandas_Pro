## Search

```python
mystr = """Batman number 456-789-2222, his Batmobile number 456-123-5555, Batwoman number 258-7777.
Batbad say: Ha, HaHa, HaHaHaHaHa, HaHaHa!!!"""
numb_regex = re.compile('(\d{3})-(\d{3}-\d{4})')
number = numb_regex.search(mystr)
number.group()
```

```python
# добываем отдельно первую и вторую группу
city_numb, bat_numb = number.group(1), number.group(2), 
city_numb, bat_numb
```

```python
number.groups()
```

```python
# ? - необязательные группы и символы. 0 или 1 повторение
mystr2 = """Batwoman number 258-7777"""
numb_regex2 = re.compile('(\d{3})?-?(\d{3}-\d{4})')
number2 = numb_regex2.search(mystr2)
number2.group()
```

```python
bat_regex = re.compile('Bat(wo)?man')
number = bat_regex.search(mystr)
print(number.group())
number = bat_regex.search(mystr2)
print(number.group())
```

```python
# или
mystr = """his Batmobile number 456-123-5555, Batwoman number 258-7777.
Batbad say: Ha, HaHa, HaHaHaHaHa, HaHaHa!!!"""
bat_regex = re.compile('Bat(man|mobile|bad)')
number = bat_regex.search(mystr)
number.group(), number.group(), number.groups()
```

```python
# * - необязательные группы и символы. 0 или бесконечность повторений
wo_regex = re.compile("Bat(wo)*man")
mystr1 = """Batman search"""
mystr2= """Batwoman search"""
mystr3 = """Batwowowowoman search"""
batman = wo_regex.search(mystr1)
print(batman.group())
batman = wo_regex.search(mystr2)
print(batman.group())
batman = wo_regex.search(mystr3)
print(batman.group())
```

```python
# * - обязательные группы и символы. 1 или бесконечность повторений
wo_regex = re.compile("Bat(wo)+man")
mystr1 = """Batman search"""
mystr2= """Batwoman search"""
mystr3 = """Batwowowowoman search"""
batman = wo_regex.search(mystr1)
print(batman is None)
batman = wo_regex.search(mystr2)
print(batman.group())
batman = wo_regex.search(mystr3)
print(batman.group())
```

```python
# * - обязательные группы и символы. 1 или бесконечность повторений
wo_regex = re.compile("Bat(wo){1,}man")
mystr1 = """Batman search"""
mystr2= """Batwoman search"""
mystr3 = """Batwowowowoman search"""
batman = wo_regex.search(mystr1)
print(batman is None)
batman = wo_regex.search(mystr2)
print(batman.group())
batman = wo_regex.search(mystr3)
print(batman.group())
```

```python
# N-е количество повторений. Жадный поиск
mystr = """Batman number 456-789-2222, his Batmobile number 456-123-5555, Batwoman number 258-7777.
Batbad say: HaHaHa, Ha, HaHa, HaHaHaHaHa, HaHaHa!!!"""
numb_regex = re.compile('(Ha){1,3}')
number = numb_regex.search(mystr)
print(number.group())

numb_regex = re.compile('(Ha){4,}')
number = numb_regex.search(mystr)
print(number.group())
```

```python
# N-е количество повторений. Не жадный поиск с помощью ?
mystr = """Batman number 456-789-2222, his Batmobile number 456-123-5555, Batwoman number 258-7777.
Batbad say: HaHaHa, Ha, HaHa, HaHaHaHaHa, HaHaHa!!!"""
numb_regex = re.compile('(Ha){1,3}?')
number = numb_regex.search(mystr)
print(number.group())
```

## findall

```python
mystr = """Batman number 456-789-2222, his Batmobile number 456-123-5555, Batwoman number 258-7777.
Batbad say: Ha, HaHa, HaHaHaHaHa, HaHaHa!!!"""
numb_regex = re.compile('\d{3}-\d{3}-\d{4}')
number = numb_regex.findall(mystr)
number
```

```python
mystr = """Batman number 456-789-2222, his Batmobile number 456-123-5555, Batwoman number 258-7777.
Batbad say: Ha, HaHa, HaHaHaHaHa, HaHaHa!!!"""
numb_regex = re.compile('(\d{3})?-?(\d{3}-\d{4})')
number = numb_regex.findall(mystr)
number
```

- \d - цифра 0-9
- \D - не цифра от 0 до 9
- \w - буква, цифра или символ "_"
- \W - не \w
- \s - пробел, табуляция или новая строка
- \S - не \s
- \b - граница слова
- ^ - начало строки
- $ - конец строки
- . - любой символ

## Sub

```python
sub_regex = re.compile(r"agent \w+", re.I)
sub_regex.sub("Nowname", "My name agent Ivan, say agent Jon.")
```

```python
# замена с ссылкой на группу. Где \1 - номер группы для подстановки
agent_name_regex = re.compile(r"agent (\w)\w*", re.I)
agent_name_regex.sub(r"\1****", "Agent Jon say agent Mikle...")
```

```python
# возможно писать сложные выражения в несколько строк с комментами с помощью re.VERBOSE
phone_regex = re.compile(r"""(
                                (\d{3}|\(\d{3}\))?           # территориальный код
                                (\s|-|\.)?                   # разделитель
                                (\d{3})                      # первые 3 цифры
                                (\s|-|\.)                    # разделитель
                                (\d{4})                      # последние 4 цифры
                                (\s*(ext|x|ext.)\s*\d{2,5})? # добавочный номер
                             )""",
                         re.VERBOSE)
phone_regex.search("my number (343)-077.288 4567   extr     4568").groups()
```

```python
# Комбинации re.I, re.DOTALL, re.VERBOSE возможны, если применять "|"
# re.compile(r"", re.I | re.DOTALL)
```

- **re.ASCII** - По умолчанию \w, \W, \b, \B, \d, \D, \s, \S соответствуют все юникодные символы с соответствующим качеством. Например, \d соответствуют не только арабские цифры, но и вот такие: ٠١٢٣٤٥٦٧٨٩. re.ASCII ускоряет работу, если все соответствия лежат внутри ASCII
- **re.IGNORECASE** - Не различать заглавные и маленькие буквы. Работает медленнее, но иногда удобно
- **re.MULTILINE** - Специальные символы ^ и $ соответствуют началу и концу каждой строки
- **re.DOTALL** - По умолчанию символ \n конца строки не подходит под точку. С этим флагом точка — вообще любой символ

## **MAC адресс и скобочные группы**

### Поиск одного символа

```python
adr = "01:23:45:67:89:ab"
mac_regex = re.compile(r"[0-9a-fA-F]")
mac_regex.findall(adr)
```

### Длинный вариант решения

```python
mac_regex = re.compile(r"[0-9a-fA-F]{2}[:-][0-9a-fA-F]{2}[:-][0-9a-fA-F]{2}[:-][0-9a-fA-F]{2}[:-][0-9a-fA-F]{2}[:-][0-9a-fA-F]{2}")
mac_regex.findall(adr)
```

### search и группы с повторами

```python
mac_regex = re.compile(r"[0-9a-fA-F]{2}([:-][0-9a-fA-F]{2}){5}")
mac_regex.search(adr).group()
```

```python
list(mac_regex.finditer(adr + ' ' + adr))
```

```python
for m in mac_regex.finditer(adr + ' ' + adr):
    print(m.group())
```

### findall

```python
# но с findall так не работает
mac_regex = re.compile(r"[0-9a-fA-F]{2}([:-][0-9a-fA-F]{2}){5}")
mac_regex.findall(adr)
```

```python
# ?: - для повторов. Позволяет для группы в скобках указывать квантификаторы повтора
mac_regex = re.compile(r"[0-9a-fA-F]{2}(?:[:-][0-9a-fA-F]{2}){5}")
mac_regex.findall(adr + ' ' + adr)
```

## Еще про скобки и объект match (search, fullmatch, finditer)

```python
pattern = r'\s*([А-Яа-яЁё]+)(\d+)\s*' 
string = r'---   Опять45   ---' 
match = re.search(pattern, string) 
print(f'Найдена подстрока >{match[0]}< с позиции {match.start(0)} до {match.end(0)}') 
print(f'Группа букв >{match[1]}< с позиции {match.start(1)} до {match.end(1)}') 
print(f'Группа цифр >{match[2]}< с позиции {match.start(2)} до {match.end(2)}') 
```

### Тонкости нумерации групп

```python
pattern = r'((\d)(\d))((\d)(\d))' 
string = r'123456789' 
match = re.search(pattern, string) 
print(f'Найдена подстрока >{match[0]}< с позиции {match.start(0)} до {match.end(0)}') 
for i in range(1, 7): 
    print(f'Группа №{i} >{match[i]}< с позиции {match.start(i)} до {match.end(i)}') 
```

## Используем свои функции в регулярках

```python
def repl(m): 
    return '>censored(' + str(len(m[0])) + ')<' 
text = "Некоторые хорошие слова подозрительны: xор, хоровод, хороводоводовед." 
print(re.sub(r'\b[хХxX]\w*', repl, text))
```
